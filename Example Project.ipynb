```python
import pandas as pd

df = pd.read_excel("Book_Comprehension_Data.xlsx")

df
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

plt.plot(df["Week"], df["Class_Comprehension_Score"], marker="o")
plt.xlabel("Week")
plt.ylabel("Comprehension Score")
plt.title("Comprehension Over Time")
plt.show()
```

```python
plt.scatter(df["Book_Word_Count"], df["Class_Comprehension_Score"])
plt.xlabel("Word Count")
plt.ylabel("Comprehension Score")
plt.title("Word Count vs Comprehension")
plt.show()
```

```python
print(df.groupby("Book_Level")["Class_Comprehension_Score"].mean())
```

```python
print(df.groupby("Question_Difficulty")["Class_Comprehension_Score"].mean())
```

```python
correlation = df["Book_Word_Count"].corr(df["Class_Comprehension_Score"])
print("Correlation between word count and comprehension:", correlation)
```
